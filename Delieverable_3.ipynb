# Creating a histogram of the table:
Another diagram that creates organization. This also makes my findings easier as it reveals the area that's more common, also revealing a pattern.

```python
plt.figure(figsize=(8,5))
plt.hist(df["age"], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Heart Disease Cases")
plt.show()
```

# Creating a correlation table:

# Creating a chest pain pie chart:
The pie chart is pleasing to the eye while identifying exact values that are mostly impactful

```python
cp_labels = {
    0: "Typical Angina",
    1: "Atypical Angina",
    2: "Non-anginal Pain",
    3: "Asymptomatic"
}

cp_counts = df["cp"].map(cp_labels).value_counts()

plt.figure(figsize=(7,7))
plt.pie(cp_counts, labels=cp_counts.index, autopct="%1.1f%%")
plt.title("Chest Pain Type Distribution")
plt.show()
```
